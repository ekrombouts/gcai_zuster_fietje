```
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
```

```
# Configureer de quantisatie naar 4-bit
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,  # Schakel 4-bit quantisatie in
    bnb_4bit_use_double_quant=True,  # Gebruik dubbele quantisatie voor efficiënter geheugenbeheer
    bnb_4bit_quant_type="nf4"  # NF4 (Normal Float) wordt vaak aanbevolen voor betere prestaties
)

# Laad je getrainde model met 4-bit quantisatie
model_name = "ekrombouts/zuster_fietje3"  # Vervang door jouw modelnaam
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=quantization_config  # Gebruik de nieuwe configuratie
)

# Laad de tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```
model.push_to_hub("ekrombouts/zuster_fietje_q4", use_auth_token=True, commit_message="V2 Quantized version of ekrombouts/zuster_fietje3", private=True)
tokenizer.push_to_hub("ekrombouts/zuster_fietje_q4", use_auth_token=True, commit_message="V2 Quantized version of ekrombouts/zuster_fietje3")
```
